# Find out what percent of the NFL dataset is missing¶


```python
# Calculate total number of cells in dataframe
totalCells = np.product(df.shape)
totalCells
```

```python
# Count number of missing values  of first 10 columns
missingCount = df.isnull().sum()
missingCount[0:10].sort_values(ascending=False)
```

```python
# Calculate total number of missing values
totalMissing = missingCount.sum()
totalMissing
```

```python
# Calculate percentage of missing values
print("The NFL dataset contains", round(((totalMissing/totalCells) * 100), 2), "%", "missing values.")
```

Look at the columns down ,SideofField ,PlayTimeDiff ,TimeSecs,time from the NFL datasets.<br>.These contain missing values.<br> Which, if either, of these are missing because they don't exist? Which, if either, are missing because they weren't recorded?

```python
#calcuate missing values percent of top 10  columns
miss_percent=missingCount * 100 / df.shape[0]
miss_percent[0:10].sort_values(ascending=False)
```

<b>Reason for missing data in columns with higest nan value</b><br>The column TimeSecs,down and time i.e Number of seconds left in the game when the play was made. This means that these values are probably missing because they were not recorded, rather than because they don't exist.

# Handling missing values and their impacts 

<b>1. A Simple Option: Drop Columns with Missing Values¶</b><br> If we drop the nan values on rows then all we get the empty dataframe which indicates that there are nan values in every rows. This is not the good idea.


```python
df_null_test_row=df.dropna(axis=0)#rows
df_null_test_row.head()
```

If we drop the nan values on columns then we have only 41 columns out of 102

```python
df_null_test_col=df.dropna(axis=1)#columns
df_null_test_col.head()
```

```python
print("Columns in original dataset: %d \n" % df.shape[1])
print("Columns with na's dropped: %d" % df_null_test_col.shape[1])
```

<b>2. Imputation</b> <br>Imputation fills in the missing value with some number. The imputed value won't be exactly right in most cases, but it usually gives more accurate models than dropping the column entirely. 

```python
imputeSfPermits = df.fillna(method='ffill', axis=0).fillna("0")

imputeSfPermits.head()
```

```python
 #replace all NA's the value that comes directly after it in the same column, 
# then replace all the reamining na's with 0
df2=df.fillna(method = 'bfill', axis=0).fillna(0).head()
```
