# Find out what percent of the sf_permit dataset is missing¶


```python
# Calculate total number of cells in dataframe
totalCells = np.product(df.shape)
totalCells
```

```python
# Count number of missing values per column
missingCount = df.isnull().sum()
missingCount.sort_values(ascending=False)
```

```python
# Calculate total number of missing values
totalMissing = missingCount.sum()
totalMissing
```

```python
# Calculate percentage of missing values
print("The SF Permits dataset contains", round(((totalMissing/totalCells) * 100), 2), "%", "missing values.")
```

Look at the columns Street Number Suffix, Zipcode, Unit Suffix,Voluntary Soft-Story Retrofit and TIDF Compliance from the sf_permits datasets.<br>.These contain missing values.<br> Which, if either, of these are missing because they don't exist? Which, if either, are missing because they weren't recorded?

```python
#calcuate missing values percent of all columns
miss_percent=missingCount * 100 / df.shape[0]
miss_percent.sort_values(ascending=False)
```

<b>Reason for missing data in columns with higest nan value</b><br>1.As every address has a Zipcode, it looks like the missing values for this column are due to the values not being recorded.<br> 2.For the Street Number Suffix column, it is likely very few properties will have a suffix to the number, I see a lot of 3s, 18s, 46s, but not nearly as many 36A or 18B, so it is likely that these are missing as they don't exist.<br>3. As TIDF Compliance is new law enforced so it may be that values were not recorded <br>4.As values for Unit are not much available, the Unit Suffix values are not available<br> 5. Since Voluntary Soft-Story Retrofit is Soft story to meet earth quake regulations which is optional  so the data may not have been recorded 

# Handling missing values and their impacts 

<b>1. A Simple Option: Drop Columns with Missing Values¶</b><br> If we drop the nan values on rows then all we get the empty dataframe which indicates that there are nan values in every rows. This is not the good idea.


```python
df_null_test_row=df.dropna(axis=0)#rows
df_null_test_row.head()
```

If we drop the nan values on columns then we have only 12 columns out of 43

```python
df_null_test_col=df.dropna(axis=1)#columns
df_null_test_col.head()
```

```python
print("Columns in original dataset: %d \n" % df.shape[1])
print("Columns with na's dropped: %d" % df_null_test_col.shape[1])
```

<b>2. Imputation</b> <br>Imputation fills in the missing value with some number. The imputed value won't be exactly right in most cases, but it usually gives more accurate models than dropping the column entirely. 

```python
imputeSfPermits = df.fillna(method='ffill', axis=0).fillna("0")

imputeSfPermits.head()
```

```python
 #replace all NA's the value that comes directly after it in the same column, 
# then replace all the reamining na's with 0
df2=df.fillna(method = 'bfill', axis=0).fillna(0)
df2.head()
```
